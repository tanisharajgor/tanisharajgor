## Priority Ranking

Rank can be based on total time period damage. A few of the methods explored are:
*  Strict Column Values
*  Norm Deviation (Z-Scores)
*  Factor Weighting

```python
from numpy.lib.function_base import rot90
# Import the necessary packages/libraries
# For further processing if necessary
# import sqlite
import pandas as pd
import numpy as np
import scipy.stats as stats

# Loading the data set from google drive
rockrose = pd.read_excel('/content/LALE_client_rockrose_development_llc_11413.xlsx', header=None)

rockrose = pd.DataFrame(rockrose)

# Set the headings
rockrose.columns = rockrose.iloc[0]
rockrose = rockrose.iloc[1:]
rockrose = rockrose.reset_index(drop=True)
```

### Ranking Strictly By Column Values

Looking at temporal damage, create a prority ranking considering each of the following columns:
- 100 Yr Property Damage [MM $US]

- 100 Yr Time Element [MM $US]

- 100 Yr Downtime [Months]

- 500 Yr Property Damage [MM $US]

- 500 Yr Property Damage [MM $US]

- 500 Yr Downtime [Months]


```python
# Assigning the proper data types for columns of interest
rockrose_sec = {
    '100_Property_Damage': rockrose['100 Yr Property Damage [MM $US]'].astype(float),
    '100_Time_Element': rockrose['100 Yr Time Element [MM $US]'].astype(float),
    '100_Downtime': rockrose['100 Yr Downtime [Months]'].astype(float),
    '500_Property_Damage': rockrose['500 Yr Property Damage [MM $US]'].astype(float),
    '500_Time_Element': rockrose['500 Yr Time Element [MM $US]'].astype(float),
    '500_Downtime': rockrose['500 Yr Downtime [Months]'].astype(float)
}

rockrose = pd.DataFrame(rockrose_sec)

def generate_rank(col_identifier):
  global rockrose
  rockrose['index'] = rockrose.index

  rockrose_sorted = rockrose.sort_values(by=col_identifier, ascending=False)

  rank = rockrose_sorted.index.tolist()

  # Merge the rank col with the original df
  rockrose['rank - ' + col_identifier] = rockrose.index
  for i in range(len(rank)):
    # The index on the sheet starts at 0
    rockrose['rank - ' + col_identifier][rank[i]] = i

  rockrose = rockrose.drop('index', axis=1)
```

```python
for key in rockrose_sec:
  generate_rank(key)

rockrose

# Sort the output by specific rank/column identifier. (This can be changed)
rockrose.sort_values(by='rank - ' + list(rockrose_sec.keys())[0], ascending=True)
```

### Ranking by Measure of Deviance

In addition, it may beneficial to consider percentage deviance as an additional method of looking at the temporal damage.

For each of the columns mentioned above, z-scores will be calculated based on the mean and standard deviations of the column data.

```python
def generate_zscores(col_identifier):
  global rockrose
  rockrose['z_score - ' + col_identifier] = stats.zscore(rockrose[col_identifier])

for key in rockrose_sec:
  generate_zscores(key)

rockrose
```

```python
# To explore possible binning intervals, look at Standard Deviation, different quartiles, min/max, etc.

# One possible way of binning is z-score quartiles: 0% (min), 25%, 50%, 75%, 100% (max), where more weight
# is given to higher z-scores (i.e. greater temporal damage from the mean).

# Determine these buckets for each column.
def generate_buckets(col_identifier):
  col_statistics = rockrose['z_score - ' + col_identifier].describe()
  buckets = col_statistics[3:8]
  return buckets

col_buckets = dict()
for key in rockrose_sec:
  col_buckets[key] = generate_buckets(key)
  print(col_buckets[key])
```

### Ranking With Weights

Individual ranking analysis has been done previously for each individual column. Having said that, it is beneficial to have a set method for determining one overall priority ranking list.

1. After making the z-score bins for each column above, we can set each one to a have a certain weight factor.

2. To determine the priority ranking/score for a given data member, take the 3 factors: Property Damage, Time Element, and Downtime. Then, for a given factor, based on the z-scores generated for it, assign it a weight. Lastly, do a weighted sum for all 3 factors.

3. All data members will be sorted based on their priorty score.

These will be done separately for 100 year and 500 year timespan.

```python
# Determine the weights
def determine_col_weights(row_index):
  global rockrose
  for key in rockrose_sec:
    col_weights = []
    for row_index, row in rockrose.iterrows():
      if (rockrose.loc[row_index, "z_score - " + key] <= col_buckets[key][3]):
        weight = 0.5
      elif (rockrose.loc[row_index, "z_score - " + key] <= col_buckets[key][4]):
        weight = 0.75
      elif (rockrose.loc[row_index, "z_score - " + key] > col_buckets[key][4]):
        weight = 1.0
      col_weights.append(weight)
    rockrose['weight - ' + key] = col_weights
  return rockrose

weighted_rank_100 = []
weighted_rank_500 = []

# Find the weighted sum
for row_index, row in rockrose.iterrows():
  determine_col_weights(row_index)
  weighted_rank_100.append(
    rockrose.loc[row_index, 'weight - 100_Property_Damage'] +
    rockrose.loc[row_index, 'weight - 100_Time_Element'] +
    rockrose.loc[row_index, 'weight - 100_Downtime'])

  weighted_rank_500.append(
    rockrose.loc[row_index, 'weight - 500_Property_Damage'] +
    rockrose.loc[row_index, 'weight - 500_Time_Element'] +
    rockrose.loc[row_index, 'weight - 500_Downtime'])

rockrose['100_weighted_rank'] = weighted_rank_100
rockrose['500_weighted_rank'] = weighted_rank_500

rockrose
```

```python
# Now sort the data points by their weighted rank.
rockrose.sort_values(by='100_weighted_rank', ascending=False)[['100_weighted_rank', '500_weighted_rank']]

# The indices show a priority rank for this method, which could be combined further using other techniques or incoporating
# both 100 and 500 year timespans.
```
